```python
import pandas as pd
from sqlalchemy import create_engine
import warnings
warnings.filterwarnings('ignore')
```

```python
amazon_link = 'Finalized CSV/Amazon_movies.csv'
disney_link = 'Finalized CSV/Disney_movies.csv'
hulu_link = 'Finalized CSV/hulu_movies.csv'
netflix_link = 'Finalized CSV/netflix_movies.csv'
```

```python
amazon_df = pd.read_csv(amazon_link)
disney_df = pd.read_csv(disney_link)
hulu_df = pd.read_csv(hulu_link)
netflix_df = pd.read_csv(netflix_link)
```

```python
ama_dis = amazon_df.merge(disney_df, how='outer')
ama_dis
```

```python
ama_dis_hulu = ama_dis.merge(hulu_df, how='outer')
ama_dis_hulu
```

```python
total_merge = ama_dis_hulu.merge(netflix_df, how='outer')
total_merge
```

```python
total_merge.to_csv('Finalized CSV/total_merge.csv',index = False, header = True)
```
